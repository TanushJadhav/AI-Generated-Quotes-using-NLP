```python
import bs4
import requests
import html5lib
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://marathidigital.com/motivational-quotes-marathi/#Marathi_Motivational_Quotes"
```

```python
req = requests.get(url)
req.content
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
quotes = []
```

```python
blockquote = soup.findAll('blockquote') 
```

```python
quotes = []
for i in blockquote:
    p_tags = i.find('p')
    quotes.append(p_tags)
```

```python
quotes
```

```python
with open('quotes_1.txt', 'w', encoding='utf-8') as outfile:
    outfile.writelines((str(i)+'\n' for i in quotes))
```
