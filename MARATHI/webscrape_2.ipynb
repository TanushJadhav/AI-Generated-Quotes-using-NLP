```python
import bs4
import requests
import html5lib
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://marathijournal.in/best-motivational-quotes-in-marathi/"
```

```python
req = requests.get(url)
req.content
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
p_tag = soup.findAll('p')
p_tag
```

```python
quotes = []

for p in p_tag:
    strong_tags = p.find_all('strong')
    for s in strong_tags:
        quotes.append(s)
```

```python
len(quotes)
```

```python
quotes_db = pd.DataFrame(columns=['Quotes'])
```

```python
quotes_db['Quotes'] = quotes
```

```python
quotes_db
```

```python
quotes_db.to_csv("Marathi_quotes.csv",sep="|",encoding='utf-8')
```

```python
with open('quotes.txt', 'w', encoding='utf-8') as outfile:
    outfile.writelines((str(i)+'\n' for i in quotes))
```
